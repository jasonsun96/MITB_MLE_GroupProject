# EUR-Lex TF-IDF Baseline

Goal: build a first supervised text-classification baseline that predicts the primary `Subject_matter` from `act_raw_text`.

Pipeline mapping:
- Bronze: raw `EurLex_all.csv`
- Silver: cleaned document table with text and label
- Gold: TF-IDF features used by the classifier

Cleaning rules applied:

1. Keep only the required fields: `CELEX`, `Date_document`, `Subject_matter`, and `act_raw_text`.
2. Remove rows with missing `act_raw_text` or missing `Subject_matter`.
3. Convert `Subject_matter` into `primary_label` by taking the first semicolon-separated label.
4. Remove invalid labels such as `character(0)`.
5. Remove documents whose raw text length is below 200 characters.
6. Normalize Unicode text to ASCII to reduce encoding noise.
7. Lowercase the legal text.
8. Remove common EUR-Lex boilerplate such as `Official Journal of the European Union` and `Avis juridique important`.
9. Remove URLs, CELEX-like identifiers, dates, standalone numbers, and non-alphabetic characters.
10. Normalize repeated whitespace.
11. Remove documents whose cleaned text length is below 100 characters.

Important note: rare-class filtering is not applied in this shared cleaned file. Rare-class filtering should be done later inside each modeling experiment, depending on the model and evaluation setup.

````python
import pandas as pd

DATA_PATH = "/Users/shiyize/Desktop/EurLex_all.csv"
USECOLS = ["CELEX", "Subject_matter", "Date_document", "act_raw_text"]

# The raw file is large, so read only the columns needed for the TF-IDF baseline.
df = pd.read_csv(DATA_PATH, usecols=USECOLS)

print(df.shape)
print(df.columns.tolist())
df.head()
````

## Silver Layer: clean text and labels

Cleaning choices for the first baseline:
- Remove rows with missing full text or missing `Subject_matter`.
- Remove invalid labels such as `character(0)`.
- Use the first semicolon-separated subject matter as the primary label.
- Remove very short extracted texts, because they are usually poor OCR/scraping outputs.


````python
def get_primary_label(label):
    label = str(label)
    for sep in [";", "|", ","]:
        if sep in label:
            return label.split(sep)[0].strip()
    return label.strip()

silver = df.dropna(subset=["act_raw_text", "Subject_matter"]).copy()
silver["text"] = silver["act_raw_text"].astype(str)
silver["label"] = silver["Subject_matter"].apply(get_primary_label)

silver = silver[
    (silver["label"] != "") &
    (silver["label"].str.lower() != "character(0)") &
    (silver["text"].str.len() >= 200)
].copy()

silver = silver[["CELEX", "Date_document", "text", "label"]]

print(silver.shape)
print("Number of labels:", silver["label"].nunique())
silver.head()
````

````python
# Inspect class imbalance.
label_counts = silver["label"].value_counts()
label_counts.head(20)
````

````python
# Keep labels with enough examples to make the first baseline stable.
MIN_SAMPLES_PER_CLASS = 100

valid_labels = label_counts[label_counts >= MIN_SAMPLES_PER_CLASS].index
data_model = silver[silver["label"].isin(valid_labels)].copy()

print(data_model.shape)
print("Number of classes:", data_model["label"].nunique())
data_model["label"].value_counts().head(20)
````

## Gold Layer: TF-IDF features

For local experimentation, train on a sample first. Increase `SAMPLE_SIZE` only after the pipeline runs successfully.


````python
SAMPLE_SIZE = 30000

if len(data_model) > SAMPLE_SIZE:
    data_sample = data_model.sample(n=SAMPLE_SIZE, random_state=42).copy()
else:
    data_sample = data_model.copy()

print(data_sample.shape)
print("Number of sampled classes:", data_sample["label"].nunique())
````

````python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    data_sample["text"],
    data_sample["label"],
    test_size=0.2,
    random_state=42,
    stratify=data_sample["label"]
)

print("Train:", X_train.shape)
print("Test:", X_test.shape)
````

````python
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score, f1_score, classification_report

# LinearSVC is a strong baseline for sparse high-dimensional TF-IDF features.
tfidf_svm = Pipeline([
    ("tfidf", TfidfVectorizer(
        lowercase=True,
        stop_words="english",
        max_features=20000,
        min_df=3,
        max_df=0.85,
        ngram_range=(1, 2),
        sublinear_tf=True
    )),
    ("clf", LinearSVC(
        class_weight="balanced",
        random_state=42
    ))
])

tfidf_svm.fit(X_train, y_train)
y_pred = tfidf_svm.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("Macro F1:", f1_score(y_test, y_pred, average="macro"))
print("Weighted F1:", f1_score(y_test, y_pred, average="weighted"))
print(classification_report(y_test, y_pred))
````

````python
# Optional sanity check: inspect predictions on a few held-out documents.
results = pd.DataFrame({
    "text_preview": X_test.str.slice(0, 300).values,
    "true_label": y_test.values,
    "predicted_label": y_pred
})

results.head(10)
````

````python
# Optional: show the strongest TF-IDF/SVM terms for selected classes.
import numpy as np

vectorizer = tfidf_svm.named_steps["tfidf"]
classifier = tfidf_svm.named_steps["clf"]
feature_names = np.array(vectorizer.get_feature_names_out())

for class_name in classifier.classes_[:10]:
    class_idx = list(classifier.classes_).index(class_name)
    top_idx = np.argsort(classifier.coef_[class_idx])[-10:][::-1]
    print(f"{class_name}")
    print(", ".join(feature_names[top_idx]))
````

## Full Cleaned Data: Frequency-Filtered TF-IDF

This experiment uses the full cleaned modeling dataset instead of the earlier 30,000-row sample.

Data used:

```text
Input file: /Users/shiyize/Desktop/clean_EurLex_all.csv
Cleaned rows: 88,609
Modeling rows after keeping labels with >=100 samples: 86,910

Train/test split: stratified 80/20

```

Feature method:

```text
Frequency-filtered TF-IDF
min_df = 5
max_df = 0.75
max_features = 50,000
ngram_range = (1, 2)
```

Because full-data `LinearSVC` was too slow on this large sparse matrix, this experiment uses `SGDClassifier(loss="hinge")`, which is a scalable linear SVM-style classifier for large text classification tasks.

Result:

```text
Dummy most frequent accuracy: 0.0931, macro-F1: 0.0019
Dummy stratified accuracy:     0.0301, macro-F1: 0.0113
Dummy uniform accuracy:        0.0102, macro-F1: 0.0074

Frequency-filtered TF-IDF + SGD hinge:
Accuracy:    0.2679
Macro-F1:    0.2485
Weighted-F1: 0.2659
```

Interpretation:

This full cleaned-data result is clearly above dummy baselines, but still modest. The main causes are the fine-grained 91-class setup, overlapping legal subject categories, and the simplification of a naturally multi-label `Subject_matter` field into one primary label.


````python
# Full cleaned-data frequency-filtered TF-IDF experiment.
# This cell is left runnable for reproducibility, but the result is summarized in the markdown cell above.

import pandas as pd

from sklearn.dummy import DummyClassifier
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import SGDClassifier
from sklearn.metrics import accuracy_score, f1_score, classification_report
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline

CLEAN_DATA_PATH = "/Users/shiyize/Desktop/clean_EurLex_all.csv"
RANDOM_STATE = 42

clean_df = pd.read_csv(
    CLEAN_DATA_PATH,
    usecols=["CELEX", "Date_document", "primary_label", "clean_text"],
    keep_default_na=False
)

label_counts = clean_df["primary_label"].value_counts()
valid_labels = label_counts[label_counts >= 100].index
full_model_data = clean_df[clean_df["primary_label"].isin(valid_labels)].copy()

X_train_full, X_test_full, y_train_full, y_test_full = train_test_split(
    full_model_data["clean_text"],
    full_model_data["primary_label"],
    test_size=0.2,
    random_state=RANDOM_STATE,
    stratify=full_model_data["primary_label"]
)

full_results = []

def add_result(name, y_true, y_pred):
    full_results.append({
        "model": name,
        "accuracy": accuracy_score(y_true, y_pred),
        "macro_f1": f1_score(y_true, y_pred, average="macro"),
        "weighted_f1": f1_score(y_true, y_pred, average="weighted")
    })

for strategy in ["most_frequent", "stratified", "uniform"]:
    dummy = DummyClassifier(strategy=strategy, random_state=RANDOM_STATE)
    dummy.fit(X_train_full, y_train_full)
    add_result(f"dummy_{strategy}", y_test_full, dummy.predict(X_test_full))

frequency_filtered_tfidf = Pipeline([
    ("tfidf", TfidfVectorizer(
        lowercase=False,
        stop_words="english",
        token_pattern=r"(?u)\b[a-zA-Z][a-zA-Z]{2,}\b",
        max_features=50000,
        min_df=5,
        max_df=0.75,
        ngram_range=(1, 2),
        sublinear_tf=True,
        norm="l2"
    )),
    ("clf", SGDClassifier(
        loss="hinge",
        alpha=1e-4,
        max_iter=20,
        tol=1e-3,
        class_weight="balanced",
        random_state=RANDOM_STATE,
        n_jobs=-1
    ))
])

frequency_filtered_tfidf.fit(X_train_full, y_train_full)
y_pred_full = frequency_filtered_tfidf.predict(X_test_full)
add_result("full_clean_frequency_filtered_tfidf_sgd_hinge", y_test_full, y_pred_full)

full_results_df = pd.DataFrame(full_results)
print(full_results_df)
print(classification_report(y_test_full, y_pred_full))
````
